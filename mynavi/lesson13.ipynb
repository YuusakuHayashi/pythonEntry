```python
import pandas as pd
import re

df = pd.read_csv("data-kai.csv", encoding="SHIFT_JIS")

sumd = {}

for row in df.iterrows():
    ymd = row[1]["年月日"]
    kousui = row[1]["降水量の合計"]
    md = re.sub(r'\d{4}\/', '', ymd)
    if not (md in sumd):
        sumd[md] = 0
    sumd[md] += kousui
result = sorted(sumd.items(), key=lambda n: n[1])

top20 = result[0:20]
pd.DataFrame(top20, columns=['日付', '降水量'])
```

```python
import pandas as pd
import re

df = pd.read_csv("cities-data.csv", encoding="SHIFT_JIS")

sumd = {}

for row in df.iterrows():
    ymd = row[1]["年月日"]
    md = re.sub(r'\d+/', '', ymd)
    if not (md in sumd):
        sumd[md] = [0, 0, 0, 0]
    for i in range(1,5):
        sumd[md][i -1] += row[1][i]
        
df = pd.DataFrame.from_dict(sumd).T
df.columns = ("横浜", "箱根", "木更津", "御殿場")

df.sort_values(by="木更津", ascending=True)[0:10]
```
